#### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

### Решение
Известно:
 
 D(B) = 0.2
 
  a = 0.5 

  Для решение задачи воспользуемся формулой равномерного распределения. M(X) = a+b/2

  D(X) = (b-a)^2/12

  Имеем дисперсию и сторону а, найдем сторону в:
  
  0.2 = (b-0.5)^2/12

```
a = 0.5
D_b = 0.2

b = (D_b * 12)**0.5 + a
print('Правая величина границы В равна: ',b)

M_b = (a + b) / 2

print('Среднее значение равно: ',M_b)
```

#### 4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.

```
import scipy.stats
def norm_rasp(val):
  return scipy.stats.norm.cdf(val)
```

```
Z = (182 - 174) / 8
A = norm_rasp(Z)  
p = round((1 - A)*100,2)
print(f'а). больше 182 см равняется {p}%')

Z = (190 - 174) / 8
A = norm_rasp(Z) 
p = round((1 - A)*100,2)
print(f'б). больше 190 см равняется {p}%')

Z_B = (190 - 174) / 8 
P = norm_rasp(Z_B)
B= round(P*100,2)
Z_A = (166 - 174) / 8 
P1 = norm_rasp(Z_A)
A =  round(P1*100,2)
print(f'в). от 166 см до 190 см равняется {B-A}%')

B = (182 - 174) / 8
P = norm_rasp(B)  
B= round(P*100,2)
A = (166 - 174) / 8 
P = norm_rasp(A)
A =  round(P*100,2)
print(f'г). от 166 см до 182 см равняется {round(B-A,2)}%')

B =(190 - 174) / 8
P = norm_rasp(B) # по таблице от 2.0  
B= round(P*100,2)
A = (158 - 174) / 8
P = norm_rasp(A) 
A =  round(P*100,2)
print(f'д). от 158 см до 190 см равняется {round(B-A,2)}%')

A = (150 - 174) / 8 # - 3
P = norm_rasp(A) # по таблице от -3.0 
B = (190 - 174) / 8 # 2 
P1 = norm_rasp(B) # от 2.0
res = round(P +(1-P1),2) * 100
print(f'е). не выше 150 см или не ниже 190 см равняется {res}%')

A = (150 - 174) / 8 # - 3
P = norm_rasp(A) # по таблице от -3.0 
B = (198 - 174) / 8 # 3 
P1 = norm_rasp(B) # от 3.0
res = round(P +(1-P1) * 100,2)
print(f'ё). не выше 150 см или не ниже 198 см равняется {res}%')

B = (166 - 174) / 8 # 3 
P1 = norm_rasp(B) # от 3.0
res = round(P1 * 100,2)
print(f'ж). ниже 166 см равняется {res}%')
```
